```python
df_clean = df.copy()

df_clean["Number of ballots received by the precinct commission"] = pd.to_numeric(
    df_clean["Number of ballots received by the precinct commission"],
    errors="coerce"
)
```

```python
df_drop = df_clean.dropna(subset=[
    "Region",
    "Polling station code",
    "Total number of voters",
    "Number of voters who took part in the vote"
])
```

```python
df_fill = df_clean.copy()

df_fill["Number of voters who took part in early voting"] = \
   df_fill["Number of voters who took part in early voting"].fillna(
        df_fill["Number of voters who took part in early voting"].median()
    )

df_fill["Number of spoiled ballots"] = \
    df_fill["Number of spoiled ballots"].fillna(0) 
```

Медиана, потому что есть выбросы, 0, если отсутствие событися

```python
df_fill["City"].mode()
```

```python
city_mode = df_fill["City"].mode()[0]
df_fill["City"] = df_fill["City"].fillna(city_mode)
```

```python
num_cols = df_fill.select_dtypes(include=["int64", "float64"]).columns
num_cols

stats_df = pd.DataFrame({
    "min": df_fill[num_cols].min(),
    "max": df_fill[num_cols].max(),
    "mean": df_fill[num_cols].mean(),
    "median": df_fill[num_cols].median(),
    "mode": df_fill[num_cols].mode().iloc[0],
    "variance": df_fill[num_cols].var(),
    "skewness": df_fill[num_cols].skew(),
    "kurtosis": df_fill[num_cols].kurt()
})

stats_df
```

```python
quantiles = df_fill[num_cols].quantile([0.05, 0.25, 0.50, 0.75, 0.95])
quantiles
```

```python
df_feat = df_fill.copy()

df_feat["turnout_rate"] = (
    df_feat["Number of voters who took part in the vote"] /
    df_feat["Total number of voters"]
)

df_feat["early_voting_share"] = (
    df_feat["Number of voters who took part in early voting"] /
    df_feat["Number of voters who took part in the vote"]
)

df_feat["lukashenko_share"] = (
    df_feat["Lukashenko"] /
    df_feat["Number of voters who took part in the vote"]
)

df_feat["tsikhanouskaya_share"] = (
    df_feat["Tsikhanouskaya"] /
    df_feat["Number of voters who took part in the vote"]
)

df_feat["invalid_share"] = (
    df_feat["Number of ballots declared invalid"] /
    df_feat["Number of voters who took part in the vote"]
)

df_feat
```

```python
df_feat["station_size_group"] = pd.cut(
    df_feat["Total number of voters"],
    bins=[0, 1000, 2000, 3000, np.inf],
    labels=["small", "medium", "large", "very_large"]
)
```

```python
df_before = df_feat[["Region", "City", "station_size_group"]].head()

df_ohe = pd.get_dummies(
    df_feat,
    columns=["Region", "City", "station_size_group"],
    drop_first=True
)

df_after = df_ohe.head()
```
